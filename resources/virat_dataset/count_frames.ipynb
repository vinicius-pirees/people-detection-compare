```python
import cv2
import numpy as np
import os
import pandas as pd
import json

main_dir = os.path.abspath(os.path.dirname(os.path.dirname(os.getcwd())))
only_moving_frames = 'n'
resource_dir = os.path.join(main_dir, 'resources/virat_dataset/')


with open(os.path.join(resource_dir, 'videos_to_process.txt')) as f:
    video_list = f.read().splitlines()
```

```python
video_name_list = [video.split('.')[0] for video  in video_list]
```

```python
video_name_list 
```

```python
video_name_list.sort()
```

```python
total_frames = 0

for video_name in video_name_list:
    
    video = cv2.VideoCapture(video_name + '.mp4')

    num_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    print(video_name, "Frames:", num_frames)
    total_frames += num_frames
```

```python
groups = ['00020','01000','01010','01020','05000']
```

```python
for group in groups:
    total_frames_group = 0
    video_name_list_group = [video_name for video_name in video_name_list if video_name[8:13] == group]

    
    for video_name in video_name_list_group:

        video = cv2.VideoCapture(video_name + '.mp4')

        num_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
        total_frames_group += num_frames
    
    print("Total for group", group, ":", total_frames_group)
```
